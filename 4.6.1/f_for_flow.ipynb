```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import linregress
import pandas as pd
%matplotlib inline
```

```python
I = [26, 24, 21, 18, 14, 10]
alpha = np.cos(np.deg2rad([5, 10, 15, 20, 25, 30]))**2

fit = linregress(alpha,I)
plt.figure(figsize=(15,10))
plt.xlabel("$\cos^2 \\alpha$", fontsize=20); plt.ylabel("$I$", fontsize=20); plt.grid(True)
plt.scatter(alpha, I)
plt.plot(alpha,alpha*fit.slope+fit.intercept)
plt.savefig("pic01.png", bbox_inches='tight')
```

```python
x = np.array([3.72, 7.69, 11.8, 15.78, 19.7, 23.8])
n = np.arange(1,7)

fit = linregress(n,x)
plt.figure(figsize=(15,10))
plt.xlabel("$n$", fontsize=20); plt.ylabel("$x$", fontsize=20); plt.grid(True)
plt.scatter(n, x)
plt.plot(n,n*fit.slope+fit.intercept)
plt.savefig("pic03.png", bbox_inches='tight')
```
