```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import linregress

m = range(1,13)
data = np.array([4.58, 3.97, 3.67, 3.48, 3.38, 3.18, 3.09, 2.89, 2.68, 2.54, 2.46, 2.37])

plt.figure(figsize=(15,10))
plt.xlabel("$m$", fontsize=20); plt.ylabel("$r^2, \mu m$", fontsize=20); plt.grid(True)

d1 = abs(data[::2]-data[0])
d2 = abs(data[1::2]-data[0])

fit = linregress(m[::2],d1); plt.scatter(m[::2],d1); plt.plot(m[::2],(m[::2])*fit.slope+fit.intercept,label='темные')
fit = linregress(m[::2],d2); plt.scatter(m[::2],d2); plt.plot(m[::2],(m[::2])*fit.slope+fit.intercept,label='светлые')
plt.legend()

plt.savefig("pic.png", bbox_inches='tight')
```
