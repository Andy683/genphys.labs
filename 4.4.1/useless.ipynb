```python
import numpy as np

g = np.array([13, 13, 14, 15, 16, 16])
m = np.array([2, 59, 27, 43, 38, 42])
s = np.array([1, 37, 8, 20, 29, 36])
sins = np.array([np.sin(np.deg2rad( g[i]+m[i]*(1/60)+s[i]*(1/3600) - 1./60 )) for i in range(len(g))])
```

```python
np.round(sins,3)
```

```python
lambdas = np.array([404.7, 435.8, 491.6, 546.1, 577, 579.1])
```

```python
from matplotlib import pyplot as plt
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(lambdas,sins)
slope, intercept

plt.figure(figsize=(12,8))
plt.xlabel('$\lambda$, nm')
plt.ylabel('$\sin \phi$')
plt.plot(lambdas,slope*lambdas+intercept)
plt.scatter(lambdas,sins,c='red')
plt.grid(True)
plt.savefig("fig1.png")
plt.show()
```
